```python
#run to copy all files to the hd7copy folder
import os
import shutil

def copy_all_files(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            source_path = os.path.join(root, file_name)
            destination_path = os.path.join(destination_folder, file_name)

            # Check if the file already exists in the destination folder
            if not os.path.exists(destination_path):
                try:
                    # Copy the file to the destination folder
                    shutil.copy2(source_path, destination_path)
                    print(f"File '{file_name}' copied to destination folder.")
                except Exception as e:
                    print(f"Error copying file '{file_name}': {str(e)}")
            else:
                print(f"File '{file_name}' already exists in the destination folder. Skipped.")

# Specify the source and destination folders
hd7_folder_path = '/content/drive/MyDrive/HD7'
destination_folder = '/content/drive/MyDrive/hd7copy'

# Call the function to copy all files
copy_all_files(hd7_folder_path, destination_folder)
```

```python
# run to count amount of files in a folder
import os

def count_files_in_folder(folder_path):
    # Initialize a counter for the number of files
    file_count = 0

    # Iterate through all files in the folder
    for _, _, files in os.walk(folder_path):
        file_count += len(files)

    return file_count

# Specify the folder path
folder_path = '/content/drive/MyDrive/HD7'

# Call the function to count files
number_of_files = count_files_in_folder(folder_path)

# Print the result
print(f"The number of files in the folder '{folder_path}' is: {number_of_files}")
```

```python
import os
import zipfile

def unzip_all_files(source_folder, destination_folder, count):
    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)

            # Check if the file is a zip file
            if file_name.lower().endswith('.zip'):
                # Create a folder with the same name as the zip file (without the extension)
                extraction_folder_name = file_name.replace('.zip', '')
                extraction_folder_path = os.path.join(destination_folder, extraction_folder_name)

                # Check if the extraction folder already exists
                if not os.path.exists(extraction_folder_path):
                    os.makedirs(extraction_folder_path)

                    # Unzip the file to the destination folder
                    with zipfile.ZipFile(file_path, 'r') as zip_ref:
                        zip_ref.extractall(extraction_folder_path)
                        count = count + 1
                        print(f'{count} Files extracted from \'{file_name}\' to \'{extraction_folder_name}\'.')
                else:
                  count = count + 1
                  print(f"Skip: '{count} {file_name}' has already been extracted to \'{extraction_folder_name}\'.")

# Specify the source and destination folders
source_folder_path = '/content/drive/MyDrive/hd7copy'
destination_folder_path = '/content/drive/MyDrive/hd7unzipped'

# Initialize count outside the function
count = 0

# Call the function to unzip all files
unzip_all_files(source_folder_path, destination_folder_path, count)
```

```python
import os
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.http import MediaFileUpload
from google.auth.transport.requests import Request
import zipfile
import requests

# Set up the Google Drive API
SCOPES = ['https://www.googleapis.com/auth/drive.file']

def authenticate_google_drive():
    creds = None
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json')
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file('path/to/client_secrets.json', SCOPES)
            creds = flow.run_local_server(port=0)
        with open('token.json', 'w') as token:
            token.write(creds.to_json())
    return creds

def build_drive_service():
    creds = authenticate_google_drive()
    return build('drive', 'v3', credentials=creds)

def download_interiornet_images(download_directory):
    # Replace 'url_to_interiornet_data' with the actual URL to the InteriorNet dataset
    url_to_interiornet_data = 'https://example.com/interiornet_data.zip'

    # Download the InteriorNet dataset
    response = requests.get(url_to_interiornet_data)

    # Extract the zip file
    with zipfile.ZipFile(BytesIO(response.content), 'r') as zip_ref:
        zip_ref.extractall(download_directory)

def get_cam0_image_paths(directory):
    cam0_image_paths = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.startswith("Cam0"):
                cam0_image_paths.append(os.path.join(root, file))
    return cam0_image_paths

def upload_images_to_drive(drive_service, upload_folder_id, image_paths):
    for image_path in image_paths:
        file_metadata = {
            'name': os.path.basename(image_path),
            'parents': [upload_folder_id]
        }
        media = MediaFileUpload(image_path, resumable=True)
        drive_service.files().create(body=file_metadata, media_body=media, fields='id').execute()

if __name__ == "__main__":
    upload_folder_id = 'your_upload_folder_id'
    download_directory = 'interiornet_images'

    # Download all images from InteriorNet
    download_interiornet_images(download_directory)

    # Get paths of all Cam0 images
    cam0_image_paths = get_cam0_image_paths(download_directory)

    # Set up the Google Drive service
    drive_service = build_drive_service()

    # Upload Cam0 images to Google Drive
    upload_images_to_drive(drive_service, upload_folder_id, cam0_image_paths)

    print(f"All Cam0 images downloaded from InteriorNet and uploaded to Google Drive.")
```
