```python
#run to copy all files to the hd7copy folder
import os
import shutil

def copy_all_files(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            source_path = os.path.join(root, file_name)
            destination_path = os.path.join(destination_folder, file_name)

            # Check if the file already exists in the destination folder
            if not os.path.exists(destination_path):
                try:
                    # Copy the file to the destination folder
                    shutil.copy2(source_path, destination_path)
                    print(f"File '{file_name}' copied to destination folder.")
                except Exception as e:
                    print(f"Error copying file '{file_name}': {str(e)}")
            else:
                print(f"File '{file_name}' already exists in the destination folder. Skipped.")

# Specify the source and destination folders
hd7_folder_path = '/content/drive/MyDrive/HD7'
destination_folder = '/content/drive/MyDrive/hd7copy'

# Call the function to copy all files
copy_all_files(hd7_folder_path, destination_folder)
```

```python
# run to count amount of files in a folder
import os

def count_files_in_folder(folder_path):
    # Initialize a counter for the number of files
    file_count = 0

    # Iterate through all files in the folder
    for _, _, files in os.walk(folder_path):
        file_count += len(files)

    return file_count

# Specify the folder path
folder_path = '/content/drive/MyDrive/HD7'

# Call the function to count files
number_of_files = count_files_in_folder(folder_path)

# Print the result
print(f"The number of files in the folder '{folder_path}' is: {number_of_files}")
```

```python
import os
import zipfile

def unzip_all_files(source_folder, destination_folder, count):
    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)

            # Check if the file is a zip file
            if file_name.lower().endswith('.zip'):
                # Create a folder with the same name as the zip file (without the extension)
                extraction_folder_name = file_name.replace('.zip', '')
                extraction_folder_path = os.path.join(destination_folder, extraction_folder_name)

                # Check if the extraction folder already exists
                if not os.path.exists(extraction_folder_path):
                    os.makedirs(extraction_folder_path)

                    # Unzip the file to the destination folder
                    with zipfile.ZipFile(file_path, 'r') as zip_ref:
                        zip_ref.extractall(extraction_folder_path)
                        count = count + 1
                        print(f'{count} Files extracted from \'{file_name}\' to \'{extraction_folder_name}\'.')
                else:
                  count = count + 1
                  print(f"Skip: '{count} {file_name}' has already been extracted to \'{extraction_folder_name}\'.")

# Specify the source and destination folders
source_folder_path = '/content/drive/MyDrive/hd7copy'
destination_folder_path = '/content/drive/MyDrive/hd7unzipped'

# Initialize count outside the function
count = 0

# Call the function to unzip all files
unzip_all_files(source_folder_path, destination_folder_path, count)
```
