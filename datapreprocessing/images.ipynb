```python
#run to copy all files to the hd7copy folder
import os
import shutil

def copy_all_files(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            source_path = os.path.join(root, file_name)
            destination_path = os.path.join(destination_folder, file_name)

            # Check if the file already exists in the destination folder
            if not os.path.exists(destination_path):
                try:
                    # Copy the file to the destination folder
                    shutil.copy2(source_path, destination_path)
                    print(f"File '{file_name}' copied to destination folder.")
                except Exception as e:
                    print(f"Error copying file '{file_name}': {str(e)}")
            else:
                print(f"File '{file_name}' already exists in the destination folder. Skipped.")

# Specify the source and destination folders
hd7_folder_path = '/content/drive/MyDrive/HD7'
destination_folder = '/content/drive/MyDrive/hd7copy'

# Call the function to copy all files
copy_all_files(hd7_folder_path, destination_folder)
```

```python
# run to count amount of files in a folder
import os

def count_files_in_folder(folder_path):
    # Initialize a counter for the number of files
    file_count = 0

    # Iterate through all files in the folder
    for _, _, files in os.walk(folder_path):
        file_count += len(files)

    return file_count

# Specify the folder path
folder_path = '/content/drive/MyDrive/hd7_sketches'

# Call the function to count files
number_of_files = count_files_in_folder(folder_path)

# Print the result
print(f"The number of files in the folder '{folder_path}' is: {number_of_files}")
```

```python
import os

def count_folders_in_folder(folder_path):
    # Initialize a counter for the number of folders
    folder_count = 0

    # Get the list of items in the specified folder
    items = os.listdir(folder_path)

    # Iterate through the items and count only directories
    for item in items:
        item_path = os.path.join(folder_path, item)

        # Check if it's a directory
        if os.path.isdir(item_path):
            folder_count += 1

    return folder_count

# Specify the folder path
folder_path = '/content/drive/MyDrive/hd7unzipped'

# Call the function to count folders
number_of_folders = count_folders_in_folder(folder_path)

# Print the result
print(f"The number of folders in the folder '{folder_path}' is: {number_of_folders}")
```

```python
import os
import zipfile

def unzip_all_files(source_folder, destination_folder, count):
    # Iterate through all files in the source folder
    for root, dirs, files in os.walk(source_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)

            # Check if the file is a zip file
            if file_name.lower().endswith('.zip'):
                # Create a folder with the same name as the zip file (without the extension)
                extraction_folder_name = file_name.replace('.zip', '')
                extraction_folder_path = os.path.join(destination_folder, extraction_folder_name)

                # Check if the extraction folder already exists
                if not os.path.exists(extraction_folder_path):
                    os.makedirs(extraction_folder_path)

                    # Unzip the file to the destination folder
                    with zipfile.ZipFile(file_path, 'r') as zip_ref:
                        zip_ref.extractall(extraction_folder_path)
                        count = count + 1
                        print(f'{count} Files extracted from \'{file_name}\' to \'{extraction_folder_name}\'.')
                else:
                  count = count + 1
                  print(f"Skip: '{count} {file_name}' has already been extracted to \'{extraction_folder_name}\'.")

# Specify the source and destination folders
source_folder_path = '/content/drive/MyDrive/hd7copy'
destination_folder_path = '/content/drive/MyDrive/hd7unzipped'

# Initialize count outside the function
count = 0

# Call the function to unzip all files
unzip_all_files(source_folder_path, destination_folder_path, count)
```

```python
import os
import shutil

def copy_all_images_to_destination(root_folder, destination_folder):
    # Ensure the destination folder exists
    os.makedirs(destination_folder, exist_ok=True)

    # Initialize a counter for naming the copied images
    counter = 1

    # Iterate through all subfolders in the root folder recursively
    for root, dirs, files in os.walk(root_folder):
        # Check if the current folder contains a 'cam0/data' subfolder
        if 'cam0' in dirs and 'data' in os.listdir(os.path.join(root, 'cam0')):
            data_folder = os.path.join(root, 'cam0', 'data')

            try:
                # Iterate through files in the data folder
                for filename in os.listdir(data_folder):
                    file_path = os.path.join(data_folder, filename)

                    # Check if it's a file
                    if os.path.isfile(file_path):
                        # Generate a new name for the file
                        new_filename = f'{counter}.png'
                        counter += 1

                        # Copy the file to the destination folder with the new name
                        destination_path = os.path.join(destination_folder, new_filename)
                        shutil.copyfile(file_path, destination_path)
                        print(f'Copied image: {file_path} to {destination_path}')

            except Exception as e:
                print(f"Error processing folder: {data_folder}")
                print(f"Error details: {e}")

# Specify the root folder containing subfolders
root_folder = '/content/drive/MyDrive/hd7unzipped'

# Specify the destination folder
destination_folder = '/content/drive/MyDrive/dataset_images'

# Call the function to copy all images with numerical names
copy_all_images_to_destination(root_folder, destination_folder)
```

```python
import os
import shutil

def get_last_index(destination_folder):
    # Get the last numerical name from existing files in the destination folder
    existing_files = [f for f in os.listdir(destination_folder) if f.endswith('.png')]
    numerical_names = [int(os.path.splitext(file)[0]) for file in existing_files]
    return max(numerical_names, default=0)

def get_last_index_from_checkpoint(checkpoint_file):
    # Read the last successfully copied index from a file
    try:
        with open(checkpoint_file, 'r') as file:
            return int(file.read())
    except (FileNotFoundError, ValueError):
        return 0

def save_last_index_to_checkpoint(checkpoint_file, index):
    # Save the last successfully copied index to a file
    with open(checkpoint_file, 'w') as file:
        file.write(str(index))

def copy_all_images_to_destination(root_folder, destination_folder, checkpoint_file):
    # Ensure the destination folder exists
    os.makedirs(destination_folder, exist_ok=True)

    # Get the last numerical name in the destination folder
    last_index = get_last_index(destination_folder)

    # Get the last successfully copied index from the checkpoint file
    last_index_checkpoint = get_last_index_from_checkpoint(checkpoint_file)

    # Determine the starting index as the maximum of the two
    starting_index = max(last_index, last_index_checkpoint) + 1

    # Initialize a counter for naming the copied images
    counter = starting_index

    # Iterate through all subfolders in the root folder recursively
    for root, dirs, files in os.walk(root_folder):
        # Check if the current folder contains a 'cam0/data' subfolder
        if 'cam0' in dirs and 'data' in os.listdir(os.path.join(root, 'cam0')):
            data_folder = os.path.join(root, 'cam0', 'data')

            try:
                # Iterate through files in the data folder
                for filename in sorted(os.listdir(data_folder)):
                    file_path = os.path.join(data_folder, filename)

                    # Check if it's a file
                    if os.path.isfile(file_path):
                        # Generate a new name for the file
                        new_filename = f'{counter}.png'
                        counter += 1

                        # Copy the file to the destination folder with the new name
                        destination_path = os.path.join(destination_folder, new_filename)
                        shutil.copyfile(file_path, destination_path)
                        print(f'Copied image: {file_path} to {destination_path}')

            except Exception as e:
                print(f"Error processing folder: {data_folder}")
                print(f"Error details: {e}")
                break  # Stop processing folders if an error occurs

    # Save the last successfully copied index to the checkpoint file
    save_last_index_to_checkpoint(checkpoint_file, counter - 1)

# Specify the root folder containing subfolders
root_folder = '/content/drive/MyDrive/hd7unzipped'

# Specify the destination folder
destination_folder = '/content/drive/MyDrive/hd7images'

# Specify the checkpoint file
checkpoint_file = '/content/drive/MyDrive/lastcheckpoint.txt'

# Call the function to copy all images with numerical names
copy_all_images_to_destination(root_folder, destination_folder, checkpoint_file)
```

```python
#image processing
import cv2
import numpy as np
from pathlib import Path
import os
import shutil

def convert_to_sketch(image_path):
    print(f"Reading image: {image_path}")
    img = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
    print("Image read successfully")

    print("Applying GaussianBlur")
    img_blur = cv2.GaussianBlur(img, (5, 5), 0)
    print("GaussianBlur applied successfully")

    print("Applying Canny edge detection")
    edges = cv2.Canny(img_blur, 30, 100)
    print("Canny edge detection applied successfully")

    print("Inverting the image")
    edges_inv = cv2.bitwise_not(edges)
    print("Image inverted successfully")

    return edges_inv

def save_sketch(image_path, output_folder):
    # Get the filename without extension
    filename = Path(image_path).stem

    # Convert the image to a sketch
    sketch = convert_to_sketch(image_path)

    # Save the sketch to the output folder with the original filename
    output_path = Path(output_folder) / f"{filename}_sketch.png"
    cv2.imwrite(str(output_path), sketch)

def process_images(image_folder, output_folder, checkpoint_file=None):
    # Ensure the output folder exists
    Path(output_folder).mkdir(parents=True, exist_ok=True)

    # Get a list of all PNG files in the input folder
    image_paths = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.lower().endswith('.png')]

    # Load checkpoint
    processed_images = set()
    if checkpoint_file:
        with open(checkpoint_file, 'r') as f:
            processed_images = set(line.strip() for line in f)

    for image_path in image_paths:
        # Skip already processed images
        if image_path in processed_images:
            print(f"Skipped {image_path} - Already processed")
            continue

        try:
            print(f"Processing {image_path}")
            save_sketch(image_path, output_folder)
            # Update checkpoint
            with open(checkpoint_file, 'a') as f:
                f.write(f"{image_path}\n")
            print(f"Processed {image_path} - Sketch saved successfully")
        except Exception as e:
            print(f"Error processing image {image_path}: {e}")

# Input and output folders
image_folder = '/content/drive/MyDrive/hd7images'
output_folder = '/content/drive/MyDrive/hd7_sketches'

# Checkpoint file
checkpoint_file = '/content/drive/MyDrive/checkpointsketch.txt'

# Process images
process_images(image_folder, output_folder, checkpoint_file)
```

```python
# image processing
import cv2
import numpy as np
from pathlib import Path
import os
counter=0
def convert_to_sketch(image_path):
    print(f"Reading image: {image_path}")
    img = cv2.imread(str(image_path), cv2.IMREAD_GRAYSCALE)
    print("Image read successfully")

    print("Applying GaussianBlur")
    img_blur = cv2.GaussianBlur(img, (5, 5), 0)
    print("GaussianBlur applied successfully")

    print("Applying Canny edge detection")
    edges = cv2.Canny(img_blur, 30, 100)
    print("Canny edge detection applied successfully")

    print("Inverting the image")
    edges_inv = cv2.bitwise_not(edges)
    print("Image inverted successfully")

    return edges_inv

def save_sketch(image_path, output_folder):
    # Get the filename without extension
    filename = Path(image_path).stem

    # Check if the sketch already exists in the output folder
    output_path = Path(output_folder) / f"{filename}_sketch.png"
    if output_path.is_file():
        print(f"Skipped {image_path} - Sketch already exists in the output folder")
        return

    # Convert the image to a sketch
    sketch = convert_to_sketch(image_path)

    # Save the sketch to the output folder with the original filename
    cv2.imwrite(str(output_path), sketch)
    print(f"Processed {image_path} - Sketch saved successfully ",counter)

def process_images(image_folder, output_folder, checkpoint_file=None):
    # Ensure the output folder exists
    Path(output_folder).mkdir(parents=True, exist_ok=True)

    # Get a list of all PNG files in the input folder
    try:
        image_paths = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.lower().endswith('.jpg')]
    except Exception as e:
        print(f"Error listing files in {image_folder}: {e}")
        # Ensure image_paths is defined even in case of an exception
        image_paths = []

    # Load checkpoint
    processed_images = set()
    if checkpoint_file:
        with open(checkpoint_file, 'r') as f:
            processed_images = set(line.strip() for line in f)

    # Iterate through all image paths
    for image_path in image_paths:
        # Skip already processed images
        if image_path in processed_images:
            print(f"Skipped {image_path} - Already processed")
            continue

        try:
            print(f"Processing {image_path}")
            save_sketch(image_path, output_folder)
            # Update checkpoint
            if checkpoint_file:
                with open(checkpoint_file, 'a') as f:
                    f.write(f"{image_path}\n")
        except Exception as e:
            print(f"Error processing image {image_path}: {e}")

# Input and output folders
image_folder = '/content/drive/MyDrive/testganffn'
output_folder = '/content/drive/MyDrive/testGANffnsketch'

# Checkpoint file
checkpoint_file = '/content/drive/MyDrive/GANfnncheckpoint.txt'

# Process all images in the input folder
process_images(image_folder, output_folder, checkpoint_file)
```
